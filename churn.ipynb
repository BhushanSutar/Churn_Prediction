```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
import pickle
```

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Churn_Modelling.csv")
```

```python
df.head()
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df.head()
```

#### Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
```

```python
df['Geography'].unique()
```

```python
Ohe=OneHotEncoder()
Encoded=Ohe.fit_transform(df[['Geography']]).toarray()
```

```python
geo_df = pd.DataFrame(Encoded, columns=Ohe.get_feature_names_out(['Geography']))
```

```python
df=pd.concat([df.drop('Geography',axis=1),geo_df],axis=1)
```

```python
df.head()
```

```python
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
df
```

```python
x=df.drop('Exited',axis=1)
y=df['Exited']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    Dropout(0.3),   # 30% neurons dropped
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop=EarlyStopping(monitor='val_loss', patience=25, restore_best_weights=True,verbose=1)
```

```python
history=model.fit(x_train,y_train,epochs=100,batch_size=32,validation_data=(x_test,y_test),callbacks=[early_stop])
```

```python
import pickle
```

```python
model.save('Churn_Modelling.h5')
```
